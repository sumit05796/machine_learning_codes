```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the speech.txt file and create a word list
file_name = "speech.txt"
word_list = []

with open('speech.txt','r') as f:
    word = f.read()
    content=word.lower()
    word_list = content.split(",")
    f.close()
    
```

```python
#Read 'stopwords.csv' file as pandas dataframe
df = pd.read_csv("stopwords.csv",index_col="Language")
```

```python
eng = df.loc[["English"]]
```

```python
stopwords_list = eng.values.tolist()
```

```python
#Removing stopwords from the word list extracted from the play

pruned_list = [i for i in word_list if i not in stopwords_list]
```

```python
### edTest(test_count) ###
#Creating a frequency dictionary with words as keys and count of their occurrences as values 

frequency_dict = {}

for i in pruned_list:
    if i not in frequency_dict:
        frequency_dict[i] = 1
    else:
        frequency_dict[i]+= 1
```

```python
#Sort the dictionary based on highest frequncy of occurence

frequncy_list = sorted(frequency_dict.items(), key = lambda x: -x[1])
```

```python
#Separate the frequency_list into 2 lists for easy plotting

words = [i[0] for i in frequncy_list]
frequency = [i[1] for i in frequncy_list]
```

```python
#Select the top 15 words to plot
x = words[0:15]
y = frequency[0:15]
```

```python
#Plotting the words on the x-axis and frequency along the y-axis

fig = plt.figure()
plt.bar(x,y,width=5)
plt.xlabel(word)
plt.ylabel(frequency)
plt.show()
```
